```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv("../data/synthetic/puntos_reciclaje_consol.csv", encoding="latin1")
```

```python
geolocator = Nominatim(user_agent="ecolima_geocode")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def buscar_coord(row):
    direccion = str(row["Direccion_Completa"]) + ", " + str(row["Distrito"]) + ", Lima, Perú"
    loc = geocode(direccion)
    if loc:
        return pd.Series([loc.latitude, loc.longitude])
    return pd.Series([None, None])
```

```python
df[["latitude", "longitude"]] = df.apply(buscar_coord, axis=1)
```

```python
df.to_csv("puntos_reciclaje_geocodificado.csv", index=False)
```

```python
# Cargar el archivo geocodificado en un nuevo DataFrame y resumir resultados
geocoded_df = pd.read_csv("../data/puntos_reciclaje_geocodificado.csv")
```

```python
# Contar registros con latitud y longitud válidas
valid_mask = geocoded_df["latitude"].notna() & geocoded_df["longitude"].notna()
valid_count = valid_mask.sum()
missing_count = len(geocoded_df) - valid_count
valid_pct = valid_count / len(geocoded_df) * 100 if len(geocoded_df) > 0 else 0

print(f"Total registros: {len(geocoded_df)}")
print(f"Con latitud y longitud: {valid_count}")
print(f"Sin latitud o longitud: {missing_count}")
print(f"Porcentaje con coordenadas: {valid_pct:.2f}%")

geocoded_df.head()
```
