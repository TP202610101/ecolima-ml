# Analisis de correlacion contra `is_optimal` - dataset real v1Este notebook usa el dataset real actual y calcula correlaciones contra el target binario `is_optimal`.> Nota: es un analisis exploratorio inicial. El dataset tiene pocas filas y esta desbalanceado por distrito.

## 1. Cargar librerias y rutas

```python
from pathlib import Pathimport numpy as npimport pandas as pdimport matplotlib.pyplot as pltimport seaborn as snsROOT = Path(r"C:/Users/nikole.garcia/Repositorios/TESIS/ecolima-ml")DATA_PATH = ROOT / "data" / "real" / "dataset_entrenamiento_v1.csv"OUT_DIR = ROOT / "outputs" / "eda"OUT_DIR.mkdir(parents=True, exist_ok=True)TARGET = "is_optimal"DATA_PATH
```

## 2. Cargar dataset

```python
df_original = pd.read_csv(DATA_PATH)df = df_original.copy()print("Shape original:", df.shape)display(df.head())
```

## 3. Materializar features derivadasEl pipeline de entrenamiento calcula algunas features automaticamente con `FeatureEngineer`. Para EDA/correlacion conviene tenerlas como columnas visibles.

```python
if "poi_parks_500m" not in df.columns and "has_park_300m" in df.columns:    df["poi_parks_500m"] = df["has_park_300m"].astype(float)if "dist_nearest_recycling_m" in df.columns:    df["recycling_deficit"] = (df["dist_nearest_recycling_m"] >= 500).astype(int)if "coverage_gap_index" not in df.columns and "recycling_deficit" in df.columns:    df["coverage_gap_index"] = df["recycling_deficit"].astype(float)if "walkability_score" in df.columns and "dist_nearest_road_m" in df.columns:    road_score = 1 - (df["dist_nearest_road_m"] / 300).clip(0, 1)    df["accessibility_composite"] = 0.6 * df["walkability_score"] + 0.4 * road_scoreif "nse_ab_pct" in df.columns:    df["nse_high_ratio"] = df["nse_ab_pct"]MATERIALIZED_PATH = ROOT / "data" / "real" / "dataset_entrenamiento_v1_materialized.csv"df.to_csv(MATERIALIZED_PATH, index=False)print("Shape materializado:", df.shape)print("Guardado en:", MATERIALIZED_PATH)
```

## 4. Balance del target y por distrito

```python
display(df[TARGET].value_counts(dropna=False).rename_axis(TARGET).reset_index(name="count"))print(f"Tasa positiva: {df[TARGET].mean():.2%}")district_col = "Distrito" if "Distrito" in df.columns else "district_id"display(pd.crosstab(df[district_col], df[TARGET]))
```

## 5. Calcular correlaciones numericas contra el targetComo el target es binario, la correlacion de Pearson contra `is_optimal` funciona como correlacion point-biserial.

```python
numeric = df.select_dtypes(include=[np.number]).copy()rows = []for col in numeric.columns:    if col == TARGET:        continue    pair = numeric[[col, TARGET]].dropna()    if len(pair) < 3 or pair[col].nunique(dropna=True) < 2:        pearson = np.nan        spearman = np.nan    else:        pearson = pair[col].corr(pair[TARGET], method="pearson")        spearman = pair[col].corr(pair[TARGET], method="spearman")    rows.append({        "feature": col,        "pearson_vs_target": pearson,        "spearman_vs_target": spearman,        "abs_pearson": abs(pearson) if pd.notna(pearson) else np.nan,        "n_valid": len(pair),        "missing": int(numeric[col].isna().sum()),        "unique_values": int(numeric[col].nunique(dropna=True)),        "mean_when_target_0": pair.loc[pair[TARGET] == 0, col].mean() if len(pair) else np.nan,        "mean_when_target_1": pair.loc[pair[TARGET] == 1, col].mean() if len(pair) else np.nan,    })corr = pd.DataFrame(rows).sort_values("abs_pearson", ascending=False)CORR_CSV = OUT_DIR / "correlation_target_real_v1.csv"corr.to_csv(CORR_CSV, index=False)display(corr.head(20))print("CSV guardado en:", CORR_CSV)
```

## 6. Grafico: top 20 correlaciones

```python
top = corr.dropna(subset=["pearson_vs_target"]).head(20).copy()TOP_PLOT = OUT_DIR / "correlation_target_real_v1_top20.png"plt.figure(figsize=(10, 7))sns.barplot(    data=top,    y="feature",    x="pearson_vs_target",    hue=top["pearson_vs_target"] > 0,    dodge=False,    palette={True: "#2F6F73", False: "#B45F4D"},)plt.axvline(0, color="black", linewidth=0.8)plt.legend([], [], frameon=False)plt.title("Top 20 correlaciones con is_optimal - dataset real v1")plt.xlabel("Correlacion Pearson / point-biserial")plt.ylabel("Feature")plt.tight_layout()plt.savefig(TOP_PLOT, dpi=180)plt.show()print("Grafico guardado en:", TOP_PLOT)
```

## 7. Matriz de correlacion: target + top 15 features

```python
MATRIX_PLOT = OUT_DIR / "correlation_matrix_real_v1_top15.png"top_features = [TARGET] + top["feature"].head(15).tolist()plt.figure(figsize=(11, 9))sns.heatmap(    numeric[top_features].corr(),    cmap="vlag",    center=0,    annot=False,    square=False,    cbar_kws={"shrink": 0.8},)plt.title("Matriz de correlacion - target y top 15 features")plt.tight_layout()plt.savefig(MATRIX_PLOT, dpi=180)plt.show()print("Matriz guardada en:", MATRIX_PLOT)
```

```python
def markdown_table(table_df, index=False):    table = table_df.copy()    if index:        table = table.reset_index()    table = table.fillna("")    headers = [str(col) for col in table.columns]    lines = [        "| " + " | ".join(headers) + " |",        "| " + " | ".join(["---"] * len(headers)) + " |",    ]    for _, row in table.iterrows():        values = [str(row[col]) for col in table.columns]        lines.append("| " + " | ".join(values) + " |")    return "\\n".join(lines)balance = df[TARGET].value_counts(dropna=False).rename_axis(TARGET).reset_index(name="count")crosstab = pd.crosstab(df[district_col], df[TARGET])show = corr[[    "feature", "pearson_vs_target", "spearman_vs_target", "abs_pearson",    "n_valid", "missing", "mean_when_target_0", "mean_when_target_1"]].head(20).copy()for col in ["pearson_vs_target", "spearman_vs_target", "abs_pearson", "mean_when_target_0", "mean_when_target_1"]:    show[col] = show[col].map(lambda x: "" if pd.isna(x) else f"{x:.4f}")REPORT_MD = OUT_DIR / "correlation_target_real_v1.md"lines = [    "# Analisis de correlacion contra is_optimal - dataset real v1",    "",    "> Generado desde notebook: 2026-06-15",    "",    "## Archivos generados",    "",    f"- Dataset materializado: `{MATERIALIZED_PATH.relative_to(ROOT)}`",    f"- Tabla CSV de correlaciones: `{CORR_CSV.relative_to(ROOT)}`",    f"- Grafico top 20: `{TOP_PLOT.relative_to(ROOT)}`",    f"- Matriz de correlacion: `{MATRIX_PLOT.relative_to(ROOT)}`",    "",    "## Dataset usado",    "",    f"- Fuente: `{DATA_PATH.relative_to(ROOT)}`",    f"- Filas: {len(df)}",    f"- Columnas originales: {df_original.shape[1]}",    f"- Columnas luego de materializar derivadas: {df.shape[1]}",    f"- Target: `{TARGET}`",    "",    "## Balance del target",    "",    markdown_table(balance),    "",    f"Tasa positiva: {df[TARGET].mean():.2%}",    "",    f"## Balance por {district_col}",    "",    markdown_table(crosstab, index=True),    "",    "## Top correlaciones con el target",    "",    markdown_table(show),    "",    "## Lectura rapida",    "",    "- La correlacion de Pearson contra un target binario equivale a una correlacion point-biserial.",    "- `recycling_density_1km` aparece entre las variables mas correlacionadas positivamente con el target.",    "- Varias variables tienen correlaciones muy parecidas porque el dataset es pequeno y esta muy condicionado por el distrito.",    "- La correlacion de `district_id` con el target confirma sesgo de distribucion entre Miraflores y Magdalena del Mar.",    "- Este analisis sirve como evidencia exploratoria inicial, pero no como conclusion fuerte del paper.",    "",    "## Limitaciones",    "",    "- Solo hay 54 filas.",    "- Solo hay 14 positivos.",    "- Miraflores tiene 4 positivos y 33 negativos; Magdalena del Mar tiene 10 positivos y 7 negativos.",    "- `centroid_lat` y `centroid_lon` tienen muchos nulos en el CSV original.",    "- Las correlaciones no prueban causalidad ni garantizan importancia predictiva en LightGBM.",]REPORT_MD.write_text("\\n".join(lines), encoding="utf-8")print("Reporte guardado en:", REPORT_MD)
```
